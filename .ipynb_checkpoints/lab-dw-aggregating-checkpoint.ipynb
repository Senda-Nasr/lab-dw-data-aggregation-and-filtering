```python
import pandas as pd
import functions
df= pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv')

df= df.drop(columns=['unnamed:_0'])  #dropping the unnamed column
display(df.duplicated().sum()) #checking  for duplicated values
df= functions.remove_full_nulls(df) #removing nan rows
df= functions.duplicate_checking(df) #removing duplicate rows
df.info()
```

```python
# 1- creating the new dataframe with total_claim_amount>1000 and respose= yes

dfnew= df[(df['total_claim_amount']>1000) & (df['response']=='Yes')]
```

```python
# 2- Analyzing the average total_claim_amount by each policy type and gender for customers who have responded "Yes" 

piv=df[(df['response']=='Yes')].pivot_table(index=['policy_type','gender'], values=['total_claim_amount'], aggfunc=['mean'])
display(piv.reset_index())
```

```python
# 3- Analyzing the total number of customers who have policies in each state
piv2= df.pivot_table(index=['state'], values=['customer'], aggfunc=['count']).reset_index()
piv2.columns=['state','customer_count']
display(piv2)

# filtering the results to only include states where there are more than 500 customers
piv2_filtered=piv2[(piv2['customer_count']>500)]
display(piv2_filtered)
```

```python
4# Find the maximum, minimum, and median customer lifetime value by education level and gender. 
piv3= df.pivot_table(index=['education','gender'], values=['customer_lifetime_value'], aggfunc=['max','min','median'])
piv3= piv3.reset_index()
piv3.columns=['education', 'gender' , 'max_customer_lifetime_value', 'min_customer_lifetime_value', 'median_customer_lifetime_value']
display(piv3)
```

```python
# 5- Analyze the number of policies sold by state and month. 
# Present the data in a table where the months are arranged as columns and the states are arranged as rows.

piv4= df.pivot_table(index=['state','month'],values=['number_of_policies'])
piv4=piv4.reset_index()
display(piv4)
piv_wide=piv4.pivot(index='state',columns='month')
display(piv_wide.reset_index())
```

```python
# 6- Display a new DataFrame that contains the number of policies sold by month, by state, for the top 3 states with the highest number of policies sold.

# creating a pivot table grouped by month and state
piv5 = df.pivot_table(index=['month','state'], values=['number_of_policies'], aggfunc=['count'])
piv5=piv5.reset_index()
piv5.columns=['month','state', 'number of policies sold']
piv5=piv5.sort_values(by='number of policies sold', ascending=False)
display(piv5)

# creating a pivot table grouped by state to find the top 3 states based on number of policies sold
top3=  df.pivot_table(index=['state'], values=['number_of_policies'], aggfunc=['count'])
top3=top3.reset_index()
top3.columns=['state', 'number of policies sold']
top3=top3.sort_values(by='number of policies sold', ascending=False)
top3=top3.reset_index()
display(top3)

# extracting list of top 3 states 
state_series= top3.loc[0:2 , 'state']
state_list= state_series.values.tolist()

# filtering the pivot table based on the list of top 3 states
print("This is the final output:")
output= piv5[piv5['state'].isin(state_list)]
display(output)
```

```python
# The marketing team wants to analyze the effect of different marketing channels on the customer response rate.
marketing_channels_piv= df.pivot_table(index=['sales_channel'], values= ['response'], aggfunc=[lambda x: (x == 'Yes').sum(), lambda x: (x == 'No').sum()])
marketing_channels_piv =marketing_channels_piv.reset_index()
marketing_channels_piv.columns=['Sales channel', 'Customers Responded Yes' , 'Customers Responded No']
display(marketing_channels_piv)
```
